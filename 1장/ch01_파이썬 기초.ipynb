```python
numbers = range(10)

for i in numbers:
    if i < 5:
        print('{}는 5보다 작습니다.'.format(i))
    elif i == 5:
        print('{}는 5와 같습니다.'.format(i))
    else:
        print('{}는 5보다 큽니다.'.format(i))
```

```python
def calculator(x, y, op):
    err = 0
    result = 0
    
    try:
        if op == '+':
            result = x + y
        elif op == '-':
            result = x - y
        elif op == '*':
            result = x * y
        elif op == '/':
            if y != 0:            
                result = x / y
        else:
            err = -1
            print('계산 실패')

        return err, result
    except:
        print('에러 발생')
        err = -1
        return err, result
```

```python
calculator(1, 2, '+')
```

```python
calculator(5, 2, '?')
```

```python
calculator('5', '2', '/')
```
