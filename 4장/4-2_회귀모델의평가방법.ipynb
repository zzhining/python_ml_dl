# 4.2 회귀 모델의 평가 방법 

## 4.2.1 MSE(Mean Squared Error)

```python
from sklearn.metrics import mean_squared_error
y_true = [0, 1, 2, 3, 4, 5]
y_pred = [0, 1.2, 2.3, 2.9, 3.8, 5.2]

# MSE 연산 결과
mean_squared_error(y_true, y_pred)
```

```python
# RMSE 연산 결과
mean_squared_error(y_true, y_pred, squared = False)
```

## 4.2.2 결정계수(R-Square)

```python
from sklearn.metrics import r2_score

y_true = [0, 1, 2, 3, 4, 5]
y_pred = [0, 1.2, 2.3, 2.9, 3.8, 5.2]
# 결정계수 연산 결과
r2_score(y_true, y_pred)
```

```python
y_true = [0, 1, 2, 3, 4, 5]
y_pred = [0, 1, 2, 3, 4, 5]
# 동일한 값을 가지는 두 데이터의 결정계수 연산 결과
r2_score(y_true, y_pred)
```

```python
y_true = [0, 1, 2, 3, 4, 5]
y_pred = [0, 1, 1, 3, 3, 4]
# 결정계수 연산 결과
r2_score(y_true, y_pred)
```
